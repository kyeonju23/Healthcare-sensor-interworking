```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('emg_210820_163000.csv', encoding = "UTF-8")
df
```

```python
ti= list(np.arange(0,10,0.01))
```

```python
df["ti"] = ti
df
```

```python
emg = df.loc[:1000,'EMG']
env = df.loc[:1000,'Envelope']
ti = df.loc[:1000,'ti']
```

```python
plt.figure(figsize = (15,10))
plt.xlabel('Time(s)',fontsize = 20), plt.ylabel('EMG signal(V)', fontsize = 20)
plt.ylim(0,3)
plt.grid()
plt.plot(ti,emg)
```

```python
plt.figure(figsize = (15,10))
plt.xlabel('Time(s)',fontsize = 20), plt.ylabel('EMG Envelope signal(V)', fontsize = 20)
plt.ylim(0,3)
plt.grid()
plt.plot(ti,env,color='g')
```

```python
plt.figure(figsize = (15,10))
plt.xlabel('Time(s)',fontsize = 20), plt.ylabel('Biometrix signal(V)', fontsize = 20)
plt.plot(ti,emg,label='EMG')
plt.plot(ti,env,label='Envelope')
plt.ylim(0,3)
plt.grid()
plt.title('Biometrix EMG / Envelope signal', fontsize = 25)
plt.legend(fontsize = 16)

#plt.savefig("./emg_signal.png")
```
